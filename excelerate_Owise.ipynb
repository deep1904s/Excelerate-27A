```python
#Most popular Opportunity category
plt.figure(figsize=(10, 5))
df["Opportunity Category"].value_counts().head(10).plot(kind="bar", color="skyblue")
plt.title("Top 10 Opportunity Categories")
plt.xlabel("Opportunity Category")
plt.ylabel("Number of Sign-Ups")
plt.xticks(rotation=45)  # Rotate labels for better visibility
plt.show()
```

```python
#opportunity sign up trend over time

df["Apply Date"] = pd.to_datetime(df["Apply Date"])
df.groupby(df["Apply Date"].dt.month)["Profile Id"].count().plot(kind="bar", figsize=(10,5))
plt.title("Monthly Sign-Up Trends")
plt.xlabel("Month")
plt.ylabel("Number of Sign-Ups")
plt.show()
```
